```
#import the necessary libraries.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

Bar Plots

```
#load dataset
tips = sns.load_dataset('tips')
sns.barplot(x ='day', y = 'tip', data = tips)
```

These black lines are called error bars.

```
sns.barplot(x = 'day', y = 'total_bill', data = tips)
```

use palette parameter to alter the colours.

```
sns.barplot(x = 'day', y = 'tip', data = tips, hue = 'sex', palette = "winter_r")
```

```
#create the pandas dataframe for easy manipulation.
df = pd.DataFrame(tips)
print(df.head())
```

```
sns.barplot(x = 'day', y = 'total_bill', data = df, hue = 'smoker')
```

interchange x and y to get a horizontal plot

```
sns.barplot(x = 'total_bill', y = 'day', data = tips, hue = "smoker", palette = "spring")
```

changing the order of the xlabels, use order

```
sns.barplot(x = 'day', y = 'tip', data = tips, order = ['Sat', 'Fri', 'Sun', 'Thur'])
```

using estimators

```
from numpy import mean, median
sns.barplot(x = 'day', y = 'total_bill', data = tips, estimator = median, palette = 'spring')
```

setting up x and y labels

```
ax = sns.barplot(x = 'day', y = 'tip', 
              data = tips, 
              palette = 'coolwarm', hue = "sex", estimator = mean)
ax.set(xlabel='day', ylabel='mean(tip)')
plt.show()
```

Using the confidence interval.

```
sns.barplot(x='smoker', y = 'tip', data  = tips, ci = 99)
```

higher the capsize longer is the horizontal width.

```
sns.barplot(x = 'day', y = 'total_bill', data = tips, palette = 'spring', capsize = 0.3)
```

Dist Plots

```
num = np.random.randn(100)
```

```
sns.distplot(num)
```

```
sns.distplot(num, color = 'red')
```

```
labels = pd.Series(num, name = 'x values')
```

```
sns.distplot(labels)
```

```
sns.distplot(labels, vertical = True)
```

```
sns.distplot(labels, vertical = True, color = 'red')
```

```
sns.distplot(labels, hist = False)
```

```
sns.distplot(labels, rug = True, hist = False)
```

Heatmaps

```
#numpy.random.randn(d0, d1, …, dn) : 
#creates an array of specified shape and fills it with random values as per standard normal distribution. 

normal = np.random.rand(12, 15)
print("Shape : ", normal.shape)
print(normal)
```

```
sns.heatmap(normal)
```

```
sns.heatmap(normal, annot = True)
```

more the range [vmin, vmax] lighter the plot gets.

```
sns.heatmap(normal, vmin = 0, vmax = 4)
```

```
#load the dataset
flights = sns.load_dataset('flights')
```

```
flights.head()
```

```
flights.tail()
```

```
flights = flights.pivot('month', 'year', 'passengers')
```

```
flights.head()
```

```
sns.heatmap(flights)
```

Visualizing correlation matrix in a better form, similar features have been coloured same.

```
sns.heatmap(flights.corr(), annot = True)
```

adding string formatting to the annotations.

```
sns.heatmap(flights, annot = True, fmt = 'd')
```

we can also add line borders

```
sns.heatmap(flights, linewidths = 0.9, annot = True, fmt = 'd')
```

```
sns.heatmap(flights, linewidths = 0.9, annot = True, fmt = 'd', vmin = 100, vmax = 300)
```

```
sns.heatmap(flights, cmap = 'RdBu', annot = True, fmt = 'd')
```

```
sns.heatmap(flights, cmap = 'summer', annot = True, fmt = 'd')
```

## If we want the distribution centered around June, 1954.

---



```
sns.heatmap(flights, cmap = 'RdBu', center = flights.loc['June', 1954], annot = True, fmt = 'd')
```

```
sns.heatmap(flights, cbar = False, cmap = 'summer',annot = True, linewidths = 1.1)
```

```
sns.heatmap(flights, cbar = False, cmap = 'summer',annot = True, fmt = 'd', linewidths = 1.1)
```

Joint plots

```
#load the two datasets.
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```
tips.head()
```

```
iris.head()
```

```
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris)
```

Draw a regression line.

```
sns.jointplot(x = 'total_bill', y = 'tip', data = tips,color = 'green', kind = 'reg')
```

```
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

Kernel Density Estimate

```
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'kde')
```

Stats functions

```
from scipy.stats import spearmanr
```

```
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, stat_func = spearmanr)
```
